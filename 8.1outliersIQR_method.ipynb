**Removing outliers using IQR method**

**Calculate the IQR (Interquartile Range):

The IQR is the range between the first quartile (Q1) and the third quartile (Q3) of the data. It represents the middle 50% of the dataset.
IQR = Q3 - Q1

Identify Outliers:
Any data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered outliers.
These are data points that are significantly higher or lower than the majority of the data.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("loan_data_set.csv")
```

```python
data.describe()
```

```python
sns.kdeplot(data["CoapplicantIncome"])
plt.show()
```

```python
q1=data["CoapplicantIncome"].quantile(0.25)
q3=data["CoapplicantIncome"].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
```

```python
min_range,max_range
```

```python
IQR_Remove_Data=data[data["CoapplicantIncome"]<=max_range]
```

**Change in orignal datset**

```python
data=IQR_Remove_Data
```

```python
sns.kdeplot(data["CoapplicantIncome"])
plt.show()
```

```python
sns.kdeplot(data["ApplicantIncome"])
plt.show()
```
