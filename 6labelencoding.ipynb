**Ordinal data, Nominal Data**

*/nominal data not in sequence,,ordinal data is in sequencenominal=labelencoding/*


  **Label encoding is typically used for nominal data, where the categories do not have any inherent order or ranking.**

**Label Encoding: Imagine you have different types of fruits like apples, bananas, and oranges. With label encoding, you'd assign each fruit a number. So, apples might be 1, bananas 2, and oranges 3. It's like giving each fruit an ID number.**

***/Fit is use to train model and transform is used to convert the model,,we can use it either seperately or together,,weuse it seperately when we do deployement of model/***

**To Make Columns**

**To Make Columns**

```python
df_city=pd.DataFrame({"city":["Jalandhar","Chandigarh","Delhi","Rajasthan","Haryana"]})
```

**To combine Columns to make one table**

```python

df_combined = pd.concat([df_name, df_city], axis=1)
df_combined
```

**import LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

**Fit is use to train model and transform is used to convert the model,,we can use it either seperately or together,,weuse it seperately when we do deployement of model**

```python
le.fit_transform(df_name["name"])
```

```python
le.fit_transform(df_city["city"])
```

**Give name to encoded column,,its individual**

```python
df_name["labelname_encoding"]=le.fit_transform(df_name["name"])
```

```python
df_name
```

```python
df_city["labelcity_encoding"]=le.fit_transform(df_city["city"])
```

```python
df_city
```

**Give name to 1 encoded column,,in table**

```python
df_combined["labelname_encoding"]=le.fit_transform(df_name["name"])
df_combined
```

**Give name to 2 encoded column,,in table**

```python
df_combined["labelcity_encoding"]=le.fit_transform(df_city["city"])
```
