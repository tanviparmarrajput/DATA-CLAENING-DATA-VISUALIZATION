```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_excel("Online Retail.xlsx")
```

```python
dataset
```

_**Total null values in each column**_

```python
dataset.isnull().sum()
```

**Total null value present in column in percentage**

```python
(dataset.isnull().sum()/(dataset.shape[0]))*100
```

**Total null data present in overall data in percentage**

```python
(dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100
```

**See Null values present through graphs**

```python
sns.heatmap(dataset.isnull())
plt.show
```

**Fill missing value by giving column name**

```python
dataset["CustomerID"].fillna(dataset["CustomerID"].mean(),inplace=True)
```

_**Total null values in each column**_

```python
dataset.isnull().sum()
```

**Fill missing value by giving column name**

```python
dataset["Description"].fillna(dataset["Description"].mode()[0],inplace=True)
```

_**Total null values in each column**_

```python
dataset.isnull().sum()
```

**After fill null values in columns overall null value in data in percentage**

```python
dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]*100)
```

**After filling values in columns overall null value in data in graph**

```python
sns.heatmap(dataset.isnull())
plt.show
```
