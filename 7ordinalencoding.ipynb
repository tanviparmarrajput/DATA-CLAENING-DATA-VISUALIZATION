**Ordinal data =  Ordinal data represents categories with a clear order or ranking. For instance, the categories "low," "medium," and "high"  or "Urban","Rural" or "Semiurban"**

```python
import pandas as pd
```

**Ordinal encoding with the help of sklearn library**

```python
df=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df
```

```python
df.head(3)
```

**Give order to data**

```python
ord_data =[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)# categories = auto(default) means encoding in 
#alphabetic order but  we want encoding according to order data so we set it according to it  
```

```python
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

**Make new column of encoded data**

```python
df["Size_en"]=oe.transform(df[["Size"]])
```

```python
df
```

**Ordinal Encoding with help of map function**

**You can assign numeric value by yourself with the help of map function**

```python
ord_data1={"s":2,"m":4,"l":6,"xl":8}
```

```python
df["Size_en_map"]=df["Size"].map(ord_data1)
```

```python
df
```
